5. Monte Carlo Method
5.1 Idea

The Monte Carlo Method uses random numbers to simulate a system and estimate a numerical result.
To estimate π: we generate random points inside a square, and count how many of those points fall inside a circle inscribed in that square.

5.2 Geometry of the Setup

Take a square of side 2r with a circle of radius r inscribed inside it (touching all four sides).
Area of the square: (2r)2=4r2(2r)^2 = 4r^2
(2r)2=4r2
Area of the circle: πr2\pi r^2

5.3 Estimating π
If points are scattered uniformly at random inside the square, the fraction of points that land inside the circle approximates the ratio of areas:
Total points inside circleTotal points≈π4\frac{\text{Total points inside circle}}{\text{Total points}} \approx \frac{\pi}{4}Total pointsTotal points inside circle​≈4π​

5.4 Algorithm

Generate a random point (x, y) where x, y ∈ [-r, r].
Check whether the point lies inside the circle: x2+y2≤r2x^2 + y^2 \le r^2
x2+y2≤r2.
Repeat for a large number of random points, keeping count of total points and points inside the circle.
Estimate π using the ratio formula above.
The more points used, the closer the estimate gets to the true value of π (see Law of Large Numbers below).

6. Law of Large Number(s)
6.1 Theory
The Law of Large Numbers (LLN) states that as the number of trials/samples in a random experiment increases, the average of the results gets closer and closer to the expected (true) value.

Example: if you find the average of random numbers between 1 and 10, the average should approach 5.5 (the true expected value) as the number of samples grows.
This is exactly why the Monte Carlo method works for estimating π: the more random points we generate, the more the ratio (points inside circle / total points) converges to the true ratio π/4.

```python
from __future__ import division
from random import random
from math import pi
import matplotlib.pyplot as plt


def rain_drop(length_of_field=1):
    """
    Simulate a random rain drop within the field.
    """
    return [(.5 - random()) * length_of_field, (.5 - random()) * length_of_field]


def is_point_in_circle(point, length_of_field=1):
    """
    Check if a point lies within the inscribed circle.
    """
    return (point[0]) ** 2 + (point[1]) ** 2 <= (length_of_field / 2) ** 2


def plot_rain_drops(drops_in_circle, drops_out_of_circle, length_of_field=1, format='png'):
    """
    Plot the raindrops in and out of the circle.
    """
    total_drops = len(drops_in_circle) + len(drops_out_of_circle)
    pi_estimate = 4 * len(drops_in_circle) / total_drops

    plt.figure()
    plt.xlim(-length_of_field / 2, length_of_field / 2)
    plt.ylim(-length_of_field / 2, length_of_field / 2)
    plt.gca().set_aspect('equal')

    plt.scatter([e[0] for e in drops_in_circle], [e[1] for e in drops_in_circle],
                color='blue', s=5, label="Drops in circle")
    plt.scatter([e[0] for e in drops_out_of_circle], [e[1] for e in drops_out_of_circle],
                color='black', s=5, label="Drops out of circle")

    plt.legend(loc="upper right")
    plt.title(f"{total_drops} drops: {len(drops_in_circle)} landed in circle, "
              f"estimating π as {pi_estimate:.4f}.")

    plt.savefig(f"rain_drops.{format}")
    plt.show()


def simulate(num_drops=5000, length_of_field=1):
    """
    Run the rain drop simulation and return the categorized drops.
    """
    drops_in_circle = []
    drops_out_of_circle = []

    for _ in range(num_drops):
        drop = rain_drop(length_of_field)
        if is_point_in_circle(drop, length_of_field):
            drops_in_circle.append(drop)
        else:
            drops_out_of_circle.append(drop)

    return drops_in_circle, drops_out_of_circle


if __name__ == "__main__":
    num_drops = 5000
    drops_in_circle, drops_out_of_circle = simulate(num_drops)

    total_drops = len(drops_in_circle) + len(drops_out_of_circle)
    pi_estimate = 4 * len(drops_in_circle) / total_drops

    print(f"Total drops: {total_drops}")
    print(f"Drops in circle: {len(drops_in_circle)}")
    print(f"Estimated pi: {pi_estimate:.6f}")
    print(f"Actual pi:    {pi:.6f}")
    print(f"Error:        {abs(pi_estimate - pi):.6f}")

    plot_rain_drops(drops_in_circle, drops_out_of_circle)
```
